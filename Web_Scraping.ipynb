# Importar as Bibliotecas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Página para fazer o scraping

https://statisticstimes.com/tech/top-computer-languages.php

# Coletar os dados da página

```python
html = requests.get("https://statisticstimes.com/tech/top-computer-languages.php").content
soup = BeautifulSoup(html, 'html5lib')
```

# Com a biblioteca Beautifulsoup vamos encontrar o primeiro parágrafo, pegar a tag P

```python
primeiro_paragrafo = soup.find('p')
primeiro_paragrafo
```

```python
primeiro_paragrafo.text
```

# Agora vamos pegar todos os parágrafos (função find_all)

```python
todos_paragrafos = soup.find_all('p')
todos_paragrafos
```

```python
todos_paragrafos[1]
```

```python
todos_paragrafos[2].text
```

# Agora vamos pegar todos os links

```python
todos_links = soup.find_all('a')
todos_links
```

# Vamos inspecionar a tabela, para pegar os dados da tebela "PYPL index (Worldwide)"

```python
tabela = soup.find('table', {'id':'table_id1'}).find('tbody')
tabela
```

```python
linhas = tabela.find_all('tr')
for linha in linhas:
  dado = linha.find_all('td')
  print(dado[0].text)
  print(dado[2].text)
  print(dado[3].text)
  print('----------')
```

# Montar uma lista de linguagens e de porcentagem com os dados, para compor um Dataframe

```python
linguagem = []
pontos = []
for linha in linhas:
  dado = linha.find_all('td')
  linguagem.append(dado[2].text)
  pontos.append(dado[3].text)
```

```python
print(linguagem)
print(pontos)
```

```python
dados = pd.DataFrame(linguagem, columns=['Linguagem'])
dados['Pontos'] = pontos
dados
```
